```python
# Dependencies and setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
# Load the exported csv from WeatherPy
city_data = pd.read_csv('../output_data/cities.csv')
city_data.head(10)
```

```python
# Drop records with missing latitude and longitude
city_data_cleaned = city_data.dropna(how='any')
city_data_cleaned.head(10)
```

```python
city_data_cleaned.dtypes
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
gmaps.configure(g_key)

# Set longtitude, latitude and humidity as the weight
coords = city_data_cleaned[["Lat","Lng"]]
weights = city_data_cleaned["Humidity"]

# Add heatmap layer to map
fig = gmaps.figure()
fig.add_layer(gmaps.heatmap_layer(coords, weights=weights))
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Set criteria for max temperature, wind speed and cloudiness and reset index
city_data_cleaned_final = city_data_cleaned.loc[(city_data_cleaned['Max Temp'] < 80) &
                                                (city_data_cleaned['Max Temp'] > 70) &
                                                (city_data_cleaned['Cloudiness'] == 0) &                                                
                                                (city_data_cleaned['Wind Speed'] < 8),:]
city_data_cleaned_final = city_data_cleaned_final.reset_index(drop=True)
city_data_cleaned_final
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Store into variable named hotel_df
hotel_df = city_data_cleaned_final

# Add a hotel name column to the dataframe
hotel_df['Hotel Name'] = ""
hotel_df
```

```python
# Use iterrows to iterate through pandas dataframe and return the location
for index, row in hotel_df.iterrows():
    
    # Get city, latitude and longtitude data from hotel_df
    latitude = row['Lat']
    longtitude = row['Lng']
    city_name = row['City']
    
    # Update location parameter
    location = f'{latitude}, {longtitude}'
    
    # Assemble url and make API request    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?keyword=hotel&radius=5000&type=loding"
    query_search = base_url + '&location=' + location + '&key=' + g_key
    response = requests.get(query_search).json()
    results = response['results']
    
    # Extract hotel results and apply try and except
    try:
        print(f"Retrieving Hotel Result for Index {index}, {city_name.title()}: {(results[0]['name']).title()}")
        hotel_df.loc[index, 'Hotel Name'] = (results[0]['name']).title()
    except (KeyError, IndexError):
        print(f"Missing result for Index {index}, {city_name}: Skipping!.")

print('------------------------------')
print('Data Retrieval Complete')
print('------------------------------')
```

```python
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(markers)

# Display figure
fig
```
